```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/SampleSuperstore.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(keep='first',inplace=True)
```

```python
data.shape
```

```python
data['Profit_Percentage'] = (data['Profit'] / data['Sales'])*100
```

```python
data.head()
```

```python
data.describe(include='all')
```

```python
sns.countplot(data=data, x='Ship Mode')
plt.show()
```

```python
plt.figure(figsize=(8,8))
data['Sub-Category'].value_counts().plot.pie(autopct='%.2f')

plt.tight_layout()
plt.show()
```

```python
sns.lineplot(data = data , x ='Discount', y='Profit')
plt.title('Profits Vs Discount')
plt.show()
```

```python
sns.lineplot(data = data , x ='Quantity', y='Profit')
plt.title('Profits Vs Quantity')
plt.show()
```

```python
best_category = data.groupby(['Category'])[['Sales', 'Quantity', 'Profit']].sum().reset_index()
best_category
```

```python
# Create a figure with three subplots in a single row
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plot the first bar plot
sns.barplot(data=best_category, x='Category', y='Sales', ax=axes[0])
axes[0].set_title('Sales')

# Plot the second bar plot
sns.barplot(data=best_category, x='Category', y='Quantity', ax=axes[1])
axes[1].set_title('Quantity')

# Plot the third bar plot
sns.barplot(data=best_category, x='Category', y='Profit', ax=axes[2])
axes[2].set_title('Profit')

# Adjust layout to prevent clipping of titles
plt.tight_layout()

# Show the plots
plt.show()
```

```python
best_sub_category = data.groupby(['Sub-Category'])[['Sales','Quantity','Profit']].sum().reset_index()
best_sub_category
```

```python
fig, axes = plt.subplots(1, 2, figsize = [15, 6])

sns.barplot(data= best_sub_category, x = 'Sub-Category', y = 'Sales', ax = axes[0])
axes[0].set_title('Sales')
axes[0].tick_params(axis='x', rotation=90)

sns.barplot(data= best_sub_category, x = 'Sub-Category', y = 'Profit', ax = axes[1])
axes[1].set_title('Profit')
axes[1].tick_params(axis='x', rotation=90)

plt.tight_layout()

plt.show()
```

```python
sns.barplot(data=best_sub_category.sort_values(by='Quantity', ascending=False), x = 'Sub-Category', y = 'Quantity')
plt.xticks(rotation = 90)
plt.show()
```

```python
best_segment = data.groupby(['Segment'])['Profit'].sum().reset_index()
best_segment
```

```python
sns.barplot(data=best_segment, x ='Segment', y = 'Profit')
plt.show()
```

```python
best_region = data.groupby(['Region'])['Profit'].sum().reset_index()
best_region
```

```python
plt.pie(best_region['Profit'],labels = best_region['Region'], autopct = '%.2f')
plt.title("Most Profitable by Region")
plt.show()
```

```python
best_city = data.groupby(['City'])['Sales'].sum().reset_index().sort_values(by='Sales', ascending=False)
top10 = best_city[:10]
bottom10 = best_city[-10:]
best_city
```

```python
fig, axes = plt.subplots(1, 2, figsize = [10,5])

sns.barplot(data = top10, x = 'City', y='Sales', ax = axes[0])
axes[0].tick_params(axis = 'x', rotation = 90)
axes[0].set_title('Top 10 City with Highest Sales')

sns.barplot(data = bottom10.sort_values(by = 'Sales', ascending = False), x = 'City', y='Sales', ax = axes[1])
axes[1].tick_params(axis = 'x', rotation = 90)
axes[1].set_title('Top 10 City with Lowest Sales')

plt.tight_layout()

plt.show()
```
